# $ ax^2 + bx + c = 0 .$


# The solution of the quadratic equation are:


# $ x_1 = \frac{-b-\sqrt{b^2-4ac}}{2a}$ and $ x_2 = \frac{-b-\sqrt{b^2-4ac}}{2a}$

```r
a  <- as.integer(readline(prompt="Please enter the value of 'a' : "))

b  <- as.integer(readline(prompt="Please enter the value of 'b' : "))

c  <- as.integer(readline(prompt="Please enter the value of 'c' : "))

#X <- as.integer(readline(prompt="Please enter the range of the graph for x: "))


determinant = b^2 -4*a*c

if (determinant == 0) { print(" The equation has one real root : ")
        print(paste("x1 = x2 ", -b/(2*a)))

} else if (determinant < 0 ) { print(" The equation has imaginary roots : " )

print(paste("x1 = ", -b/(2*a),'+',sqrt(abs(determinant))/(2*a),'i'))
print(paste("x2 = ", -b/(2*a),'-',sqrt(abs(determinant))/(2*a),'i'))


} else { print("The equation has two real roots : ")
  print(paste("X1 = ", (-b+sqrt(determinant))/(2*a)))
  print(paste("X2 = ", (-b+sqrt(determinant))/(2*a)))
}

# # ploting the graph for the equation

# x <- seq(from = -X, to = X, length.out = 101)
# #x <- seq(-15,15)

# y = a*x^2 + b*x +c

# #plot(x,y,main = r'\\y = ax^2 + bx + c ',xlab = " x - axis  ", ylab = " y axis ",  type ='l',col = 'blue')
# plot(x,y, xlab = " x - axis  ", ylab = " y axis ",  type ='l',col = 'blue')

```
